```python
def open_browser():    
    options = webdriver.ChromeOptions()
    options.add_argument("--window-size=1080,800")

    chrome_service = Service('chromedriver')
    chrome_service = Service(executable_path="chromedriver.exe")
    browser = webdriver.Chrome(service=chrome_service, options=options)
    
    return browser

def open_url(browser, url):
    browser.get(url)
    time.sleep(1)
    browser.get('http://svey.kr/ICT23/mr01.asp')
    time.sleep(1)
```

```python
def first_page_field(
    browser, s1, s2, s3, s4, s5, sw_status, sw_status_sub, ict_rd_status, 
    ict_export
):
    find_id('A8_r2', browser).click()
    time.sleep(.5)
    find_id('A9', browser).send_keys(0)
    time.sleep(.5)
    
    bn_data = find_id('A10', browser).get_attribute('value')

    if bn_data == '':
        find_id('A10', browser).send_keys('000000-0000000')
        
    time.sleep(.5)
    
    if serial_number1[0] != 0:
        find_id('A12_1', browser).send_keys(str(serial_number1[0]))
    if serial_number2[0] != 0:
        find_id('A12_2', browser).send_keys(str(serial_number2[0]))
    if serial_number3[0] != 0:
        find_id('A12_3', browser).send_keys(str(serial_number3[0]))
    if serial_number4[0] != 0:
        find_id('A12_4', browser).send_keys(str(serial_number4[0]))
    if serial_number5[0] != 0:
        find_id('A12_5', browser).send_keys(str(serial_number5[0]))
        
    find_id('A12a_1', browser).send_keys(100)
    find_id('A12a_2', browser)
    find_id('A12a_3', browser)
    find_id('A12a_4', browser)
    find_id('A12a_5', browser)
    
    time.sleep(.5)
    
    # SW 총판 여부 
    if sw_status:
        ## (Yes)
        find_xpath('/html/body/div[2]/div[2]/form/table[4]/tbody/tr/td[1]/label', browser).click()
    else:
        ## (No)
        find_xpath('/html/body/div[2]/div[2]/form/table[4]/tbody/tr/td[2]/label', browser).click()
    time.sleep(.5)
    
    # 자체개발 SW 보유여부
    if sw_status_sub:
        ## (Yes)
        find_xpath('/html/body/div[2]/div[2]/form/table[5]/tbody/tr/td[1]/label', browser).click()
    else:
        ## (No)
        find_xpath('/html/body/div[2]/div[2]/form/table[5]/tbody/tr/td[2]/label', browser).click()
    time.sleep(.5)
    
    # 연구개발(R&D)투자 및 수출
    ## ICT 연구개발(R&D)투자
    if ict_rd_status:
        ### (Yes)
        find_xpath('/html/body/div[2]/div[2]/form/table[6]/tbody/tr/td[2]/table/tbody/tr/td[1]/label', browser).click()
    else:
        ### (No)
        find_xpath('/html/body/div[2]/div[2]/form/table[6]/tbody/tr/td[2]/table/tbody/tr/td[2]/label', browser).click()
    time.sleep(.5)
    
    ## ICT수출
    if ict_export:
        ### (Yes)
        find_xpath('/html/body/div[2]/div[2]/form/table[6]/tbody/tr/td[4]/table/tbody/tr/td[1]/label[1]', browser).click()
        time.sleep(.5)
        sw_yes = find_xpath('/html/body/div[2]/div[2]/form/table[6]/tbody/tr/td[4]/table/tbody/tr/td[1]/div/label[1]', browser)
        sw_no = find_xpath('/html/body/div[2]/div[2]/form/table[6]/tbody/tr/td[4]/table/tbody/tr/td[1]/div/label[2]', browser)
        sw_no.click()
    else:
        ### (No)
        find_xpath('/html/body/div[2]/div[2]/form/table[6]/tbody/tr/td[4]/table/tbody/tr/td[1]/label[2]', browser).click()
        
    time.sleep(.5)
    
    find_id('btn-save', browser).click()

    time.sleep(1)

    try:
        alert = browser.switch_to.alert
        alert.accept()
    except:
        pass

    time.sleep(1)

    find_id('btn_submit', browser).click()
```

```python
def second_page_field(browser, a1, a2, a3, a4, a5, a6, a7, a8, a9, a10, a11, a12, a13, a14):
    # 매출액(㉠+㉡)
    ## ㉠ ICT부문 (세부 품목매출의 합)
    field1_input = find_id('A16_2', browser)
    field1_input.send_keys(a1)
    time.sleep(.5)
    ## ㉡ 기타
    field2_input = find_id('A16_3', browser)
    field2_input.send_keys(a1)
    time.sleep(.5)
    
    # ② 매출원가
    field3_input = find_id('A16_4', browser)
    field3_input.send_keys(a2)
    time.sleep(.5)

    # ③ 판매비와 관리비
    ## ㉠ 급여총액(㉡, ㉢ 제외)
    field4_input = find_id('A16_6', browser)
    field4_input.send_keys(a4 - (a5 + a6))
    time.sleep(.5)

    ## ㉡ 퇴직급여
    field5_input = find_id('A16_7', browser)
    field5_input.send_keys(a5)
    time.sleep(.5)

    ## ㉢ 복리후생비
    field6_input = find_id('A16_8', browser)
    field6_input.send_keys(a6)
    time.sleep(.5)

    ## ㉣ 임차료
    field7_input = find_id('A16_9', browser)
    field7_input.send_keys(a8)
    time.sleep(.5)

    ## ㉤ 세금과공과금
    field8_input = find_id('A16_10', browser)
    field8_input.send_keys(a7)
    time.sleep(.5)

    ## ㉥ 감가상각비(유형)
    field9_input = find_id('A16_11', browser)
    field9_input.send_keys(a9)
    time.sleep(.5)

    ## ㉦ 대손상각비
    field10_input = find_id('A16_12', browser)
    field10_input.send_keys(a11)
    time.sleep(.5)

    ## ㉧ 경상개발연구비
    field11_input = find_id('A16_13', browser)
    field11_input.send_keys(a10)
    time.sleep(.5)

    ## ㉨ 기타경비 [③ 판매비와 관리비 - ( ㉠ ~ ㉧의 합) ]
    field12_input = find_id('A16_14', browser)
    field12_input.send_keys(round(a3 - (a4 + a5 + a7 + a6 + a8 + a9 + a10 + a11), 1))
    time.sleep(.5)

    # ④ 영업이익
    field13_input = find_id('A16_15', browser)
    field13_input.send_keys(a12)
    time.sleep(.5)

    # ⑤ 영업외 수익 중 이자수익
    field14_input = find_id('A16_16', browser)
    field14_input.send_keys(a13)
    time.sleep(.5)

    # ⑤ 영업외 수익 중 이자수익
    field15_input = find_id('A16_17', browser)
    field15_input.send_keys(a14)
    time.sleep(.5)
    
    # 원가명세서 및 기타
    ## ① 급여총액(②, ③ 제외)
    find_id('A17_1', browser).send_keys(0)
    ## ② 퇴직급여
    find_id('A17_2', browser).send_keys(0)
    ## ③ 복리후생비
    find_id('A17_3', browser).send_keys(0)
    ## ④ 임차료
    find_id('A17_4', browser).send_keys(0)
    ## ⑤ 세금과공과금
    find_id('A17_5', browser).send_keys(0)
    ## ⑥ 감가상각비(유형)
    find_id('A17_6', browser).send_keys(0)
    
    time.sleep(1)
    find_id('btn-save', browser).click()

    time.sleep(1)

    try:
        alert = browser.switch_to.alert
        alert.accept()
    except:
        pass
```

```python
browser = open_browser()

sw_status = 0
sw_status_sub = 0
ict_rd_status = 0
ict_export = 0

data_list = zip(
    url_list, serial_number1, serial_number2, serial_number3, serial_number4, serial_number5, 
    aa_list1, aa_list2, aa_list3, aa_list4, aa_list5, aa_list6, aa_list7, 
    aa_list8, aa_list9, aa_list10, aa_list11, aa_list12, aa_list13, aa_list14
)

count = 1

for data in data_list:
    url, s1, s2, s3, s4, s5, a1, a2, a3, a4, a5, a6, a7, a8, a9, a10, a11, a12, a13, a14 = data
    
    open_url(browser, url)
    first_page_field(browser, s1, s2, s3, s4, s5, sw_status, sw_status_sub, ict_rd_status, ict_export)
    time.sleep(1)
    second_page_field(browser, a1, a2, a3, a4, a5, a6, a7, a8, a9, a10, a11, a12, a13, a14)
    time.sleep(2)
    print(count)
    count += 1
```
