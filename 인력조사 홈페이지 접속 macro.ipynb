```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

from webdriver_manager.chrome import ChromeDriverManager

from bs4 import BeautifulSoup as BS
from datetime import datetime

import pandas as pd
import time
import sys, os
import re
import json
import random


def find_css(css_selector, browser):
    return browser.find_element(By.CSS_SELECTOR, css_selector)
def finds_css(css_selector, browser):
    return browser.find_elements(By.CSS_SELECTOR, css_selector)

def find_xpath(xpath, browser):
    return browser.find_element(By.XPATH, xpath)
def finds_xpath(xpath, browser):
    return browser.find_elements(By.XPATH, xpath)

def find_id(e_id, browser):
    return browser.find_element(By.ID, e_id)

def find_className(cn, browser):
    return browser.find_element(By.CLASS_NAME, cn)
def finds_className(cn , browser):
    return browser.find_elements(By.CLASS_NAME, cn)

def find_linktext(lt, browser):
    return browser.find_element(By.LINK_TEXT, lt)

def find_name(name, browser):
    return browser.find_element(By.NAME, name)
def finds_name(name, browser):
    return browser.find_elements(By.NAME, name)

def find_tagName(tag_name, browser):
    return browser.find_element(By.TAG_NAME, tag_name)

def finds_tagName(tag_name, browser):
    return browser.find_elements(By.TAG_NAME, tag_name)

def resource_path(relative_path):
    """ Get absolute path to resource, works for dev and for PyInstaller """
    base_path = getattr(sys, '_MEIPASS', os.path.dirname(os.path.abspath(__file__)))
    return os.path.join(base_path, relative_path)
```

```python
def open_browser():
    options = webdriver.ChromeOptions()
    options.add_argument('--no--sandbox')
    options.add_argument('no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument('--window-size=1080,800')
    options.add_argument('incognito')
    chrome_service = Service('chromedriver')
    chrome_service = Service(executable_path="chromedriver.exe")
    browser = webdriver.Chrome(service=chrome_service, options=options)
    
    return browser
```

```python
def get_url(url, browser):
    browser.get(url)
    time.sleep(.5)
    find_className('temp_btn2', browser).click()
```

```python
def second_page_field(browser, employment_status):
    completion_data = find_id('a1_1', browser).is_selected()
    not_completed_data = find_id('a1_2', browser).is_selected()
    
    completion = find_xpath('/html/body/form/center/div/table/tbody/tr/td/table/tbody/tr/td[1]/label', browser)
    not_completed = find_xpath('/html/body/form/center/div/table/tbody/tr/td/table/tbody/tr/td[2]/label', browser)

    if not any([completion_data, not_completed_data]):
        if employment_status == 1:
            completion.click()
            
            time.sleep(1)
            find_className('temp_btn2', browser).click()
            return 1
        else:
            not_completed.click()
            
            time.sleep(1)
            find_className('temp_btn2', browser).click()
            return 0
```

```python
def third_page_field(browser, employment_number, planned_employment_number):
    field1_input = find_name('a2', browser)
    field2_input = find_name('a3', browser)

    field1_input_data = find_name('a2', browser).get_attribute('value')
    field2_input_data = find_name('a2', browser).get_attribute('value')

    # 문4.	2024년 IT/SW 인력 채용계획이 있으십니까?
    field1 = find_xpath('/html/body/form/center/div/table/tbody/tr/td/table[3]/tbody/tr[1]/td[1]/label', browser)
    field2 = find_xpath('/html/body/form/center/div/table/tbody/tr/td/table[3]/tbody/tr[1]/td[2]/label', browser)
    field3 = find_xpath('/html/body/form/center/div/table/tbody/tr/td/table[3]/tbody/tr[2]/td[1]/label', browser)
    field4 = find_xpath('/html/body/form/center/div/table/tbody/tr/td/table[3]/tbody/tr[2]/td[2]/label', browser)

    field_list = [field1, field2, field3, field4]

    if field2_input_data == '':

        # 문2.	채용을 진행하였다면, 구인한 IT/SW 인원
        field1_input.send_keys(planned_employment_number)
        time.sleep(.5)

        # 문3.	실제 채용이 완료된 IT/SW 인원
        field2_input.send_keys(employment_number)
        time.sleep(.5)

        if planned_employment_number > employment_number:
            # 채용하려고 했던 인원이 더 많을 시
            random_int = random.randint(0, 1)
            field_list[random_int].click()
        else:
            # 채용하려고 했던 인원이 더 적을 시
            random_int = random.randint(2, 3)
            field_list[random_int].click()

    time.sleep(1)
    find_className('temp_btn2', browser).click()
```
